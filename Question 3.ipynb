```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales=pd.read_csv("sales.txt",sep='\t',thousands=',')
sales['date']=pd.to_datetime(sales['date'])
sales.head(5)
```

```python
sales.dtypes
```

```python
chanA=sales.loc[sales['channel'] == 'Channel A']
chanB=sales.loc[sales['channel'] == 'Channel B']
chanC=sales.loc[sales['channel'] == 'Channel C']
```

```python
plt.figure(figsize=(20,10))
plt.plot(chanA['date'],chanA['challenger forecast'])
plt.plot(chanA['date'],chanA['champion forecast'])
plt.plot(chanA['date'],chanA['actual Sales'])
plt.legend()
plt.title('Forecast and actual sales for Channel A')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(chanB['date'],chanB['challenger forecast'])
plt.plot(chanB['date'],chanB['champion forecast'])
plt.plot(chanB['date'],chanB['actual Sales'])
plt.legend()
plt.title('Forecast and actual sales for Channel B')
plt.show()
```
